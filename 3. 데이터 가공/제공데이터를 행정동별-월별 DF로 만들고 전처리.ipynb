# 2개월간격 df를 만들기. train/test 따로 만듦

```python
#먼저 월별로 맞춰져있는 df부르기
df2= pd.read_csv('C:/Users/twentystones/Desktop/빅콘/02_평가데이터_update(210806)/지역별-월별 변수들-총합_최종_진짜_13일클러변경_클러원핫.csv', encoding='cp949')

#불필요한 칼럼 삭제
df2 = df2.drop(['emd_nm_num','emd_nm2','Unnamed: 0'], axis=1)


#2달 당기기
emd_nm2term_list =[]
for i in range(0,42):  #읍면동 수  
    #em_g 2칸씩
    df2[i*42:(i+1)*42]['em_g_interpolated'].values[0:40] = df2['em_g_interpolated'][i*42:(i+1)*42].values[2:42]
    df2[i*42:(i+1)*42]['em_g_interpolated'].values[40] = 0
    df2[i*42:(i+1)*42]['em_g_interpolated'].values[41] = 0
    
    #date 2칸씩
    df2[i*42:(i+1)*42]['date'].values[0:40] = df2['date'][i*42:(i+1)*42].values[2:42]
    df2[i*42:(i+1)*42]['date'].values[40] = pd.Timestamp(2021,7,1)
    df2[i*42:(i+1)*42]['date'].values[41] = pd.Timestamp(2021,8,1)   
    
    #emd_nm_2term칼럼 만들기
    emd = df2[i*42:(i+1)*42]['emd_nm'].values[0]
    for j in range(3,45): #월 수
        emd_nm2term_list.append(emd+str(j))
df2.insert(1, 'emd_nm_2term', emd_nm2term_list,True)

#저장
df2.to_csv('df_total_2term.csv', encoding='cp949')
df2
```

# 2달 당긴 train셋 만들기

```python
#2달 미뤄서 각 지역별로 2달씩 공백이 생김. 이걸 지우기
df2_train = df2
for i in range(1,43): #42개지역
    df2_train = df2_train.drop(index=[i*42-2, i*42-1], axis=0)   

    
#해당 지역은 EDA와 배경조사에 따를때 RFID방식의 쓰레기배출이 뒤늦게 생긴 곳이라서 20.3이후만 학습에 사용
after20 = ['애월읍','성산읍','표선면','대정읍','안덕면','조천읍','구좌읍','한림읍','한경면']

for i in after20 :
    idx = df2_train[df2_train['emd_nm']==i][0:24].index     #넉넉하게 앞에 24개 삭제. 총42-2-24=16개월만 학습에 이용
    df2_train.drop(idx, inplace=True)
    
    
#알수없음은 20.8이후만 학습에 사용
idx = df2_train[df2_train['emd_nm']=='알수없음'][0:29].index
df2_train.drop(idx, inplace=True)

#날짜순 정렬
df2_train = df2_train.sort_values(by=['date'])  

#저장
df2_train.to_csv('df2_train_최종.csv', encoding='cp949')
df2_train
```

# 2달 당긴 test셋 만들기

```python
df2_test = pd.DataFrame({}, columns=df2.columns)

#동별로 가장 최근 2달씩만 가져오기
for i in range(0,43):
    df2_test = df2_test.append(df2[i*42-2:i*42], ignore_index=False)

#날짜순 정렬
df2_test = df2_test.sort_values(by=['date'])  

#저장    
df2_test.to_csv('df2_test_최종.csv', encoding='cp949')
df2_test
```
