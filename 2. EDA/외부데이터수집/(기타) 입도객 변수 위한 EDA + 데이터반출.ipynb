```python
import pandas as pd
```

```python
# 2021년+2020년 1월~7월

kk1= pd.DataFrame()
for i in range(1,8):
    df = pd.read_excel('2021년+'+str(i)+'월관광객입도현황(확정치).xls')
    k1 = df[df['Unnamed: 5'] =='월계'].iloc[2:5,6:8]
    k1.columns = ['20210'+str(i),'20200'+str(i)]
    k1.index = ['개별여행','부분패키지','패키지']

    k2 = df[df['Unnamed: 5'] =='월계'].iloc[8:14,6:8]
    k2.columns = ['20210'+str(i),'20200'+str(i)]
    k2.index = ['레저스포츠','회의및업무','휴양및관람','친지방문','교육여행','기타방문']

    k3 = pd.concat([k1,k2])
    
    kk1 = pd.concat([kk1,k3],axis=1)
    
kk1 = kk1.T
kk1.reset_index(inplace=True)
```

```python
# 2020년+2019년 1월~9월

kk2 = pd.DataFrame()
for i in range(1,10):
    df = pd.read_excel('2020년+'+str(i)+'월관광객입도현황(확정치).xls')
    k1 = df[df['Unnamed: 5'] =='월계'].iloc[2:5,6:8]
    k1.columns = ['20200'+str(i),'20190'+str(i)]
    k1.index = ['개별여행','부분패키지','패키지']

    k2 = df[df['Unnamed: 5'] =='월계'].iloc[8:14,6:8]
    k2.columns = ['20200'+str(i),'20190'+str(i)]
    k2.index = ['레저스포츠','회의및업무','휴양및관람','친지방문','교육여행','기타방문']

    k3 = pd.concat([k1,k2])
    
    kk2 = pd.concat([kk2,k3],axis=1)

kk2 = kk2.T
kk2.reset_index(inplace=True)
```

```python
# 2020년+2019년 10월~12월

kk3 = pd.DataFrame()
for i in range(10,13):
    df = pd.read_excel('2020년+'+str(i)+'월관광객입도현황(확정치).xls')
    k1 = df[df['Unnamed: 5'] =='월계'].iloc[2:5,6:8]
    k1.columns = ['2020'+str(i),'2019'+str(i)]
    k1.index = ['개별여행','부분패키지','패키지']

    k2 = df[df['Unnamed: 5'] =='월계'].iloc[8:14,6:8]
    k2.columns = ['2020'+str(i),'2019'+str(i)]
    k2.index = ['레저스포츠','회의및업무','휴양및관람','친지방문','교육여행','기타방문']

    k3 = pd.concat([k1,k2])
    
    kk3 = pd.concat([kk3,k3],axis=1)

kk3 = kk3.T
kk3.reset_index(inplace=True)
```

```python
# 추가로 수집한 데이터 모으기
sdf = pd.concat([kk1,kk2])
sdfag = pd.concat([sdf,kk3])
sdfag.rename(columns = {'index':'dtYearMonth'},inplace=True)
sdfag
```

```python
# 기존 데이터 불러오기
plus = pd.read_csv('[임시]18-19년관광객.csv',encoding='cp949')
plus1 = pd.pivot_table(plus,index = 'dtYearMonth',columns='visitPurpose',values='visitorCnt')
plus1.reset_index(inplace=True)
```

```python
# 두개의 데이터 합치기
tt = pd.concat([sdfag,plus1])
tt.dtYearMonth = pd.to_numeric(tt.dtYearMonth)
tt = tt.sort_values(by='dtYearMonth',ascending=True)
tt = tt.drop_duplicates()
tt
```
